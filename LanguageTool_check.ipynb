### Parse to a list of Dicts

```python
def parse_prompt_response(file_path):
    with open(file_path, 'r') as file:
        content = file.read()

    pairs = content.split('---END_OF_PAIR---')
    data = []
    
    for pair in pairs:
        if '---PROMPT_START---' in pair and '---RESPONSE_START---' in pair:
            prompt, response = pair.split('---RESPONSE_START---', 1)
            prompt = prompt.split('---PROMPT_START---', 1)[-1].strip()
            response = response.strip()
            data.append({'prompt': prompt, 'response': response})
    
    return data

# Example usage
#file_path = '/Users/lahiruwijayapala/Desktop/Research Data/Case10.txt'
pairs = parse_prompt_response(file_path)
for idx, pair in enumerate(pairs):
    print(f"Prompt {idx + 1}:\n{pair['prompt']}\n")
    print(f"Response {idx + 1}:\n{pair['response']}\n")
```

```python
df=pd.DataFrame(pairs)
df.head()
```

```python
#all_prompts = "\n".join(df['prompt'].tolist())
all_prompts = "\n".join(f'"{prompt}"' for prompt in df['prompt'])

print(all_prompts)
```

```python
tool = language_tool_python.LanguageToolPublicAPI('en-US')
matches = tool.check(all_prompts)
tool.close()

print(matches)
```

```python
# Print all errors using offset and errorLength

for match in matches:
    start = match.offsetInContext
    end = start + match.errorLength
    error = match.context[start:end]
    print(f"Error: '{error}' with message:'{match.message}' position {start}-{end},{match.category}, {match.ruleIssueType}")
    
```

```python
# Prepare a list to hold error data
error_data = []

# Iterate through matches and collect data
for match in matches:
    start = match.offsetInContext
    end = start + match.errorLength
    error = match.context[start:end]
    
    error_data.append({
        "Error": error,
        "Message": match.message,
        "Category": match.category,
        "Rule Issue Type": match.ruleIssueType
    })

# Create a DataFrame from the error data
error_df = pd.DataFrame(error_data)
```

```python
from tabulate import tabulate

print(tabulate(error_df, headers='keys', tablefmt='psql'))

# This also pretty prints the DF
#print(error_df.to_markdown()) 
```
